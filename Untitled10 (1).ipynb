```python


def clean_text(text):
    text = text.lower()

    text = re.sub(r'http\S+|www.\S+', '', text)

    text = re.sub(r'[^\w\s]', '', text)

    text = re.sub(r'\d+', '', text)

    text = re.sub(r'\s+', ' ', text).strip()
    return text

df["cleaned_text"]=df["text"].apply(clean_text)
```

```python

def tokenize_and_remove_stopwords(text):
    words = word_tokenize(text)

    stop_words=set(stopwords.words("english"))

    filtered_words = [word for word in words if word not in stop_words]
    return filtered_words

df['tokens'] = df['cleaned_text'].apply(tokenize_and_remove_stopwords)

print(df["tokens"])
```

```python
def stemming_and_lemmatization(tokens):

    lemmatizer = WordNetLemmatizer()


    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]

    return  lemmatized_tokens

df["lematizer"]=df["tokens"].apply(stemming_and_lemmatization)



print(df["lematizer"])
```

I performed text cleaning to prepare the data for natural language processing. This included removing URLs, numbers, special characters, and extra white spaces. After cleaning the raw text, I tokenized the sentences into individual words using a tokenizer. Then, I removed common stopwords to reduce noise and focus on meaningful words. These preprocessing steps helped standardize the text and improve the quality of the input for the model.


```python
x=df["lematizer"]
y=df["subject"]

print(x)
print("--------------------------")
print(y)

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42,shuffle=True)

print("x_train shape",x_train.shape)
print("x_test shape",x_test.shape)
print("y_train shape",y_train.shape)
print("x_train shape",y_test.shape)

```

```python
label=LabelEncoder()

y_train=label.fit_transform(y_train)
y_test=label.transform(y_test)


print(y_train)
print("------------------")
print(y_test)
```

```python
x_train = [' '.join(tokens) for tokens in x_train]
x_test = [' '.join(tokens) for tokens in x_test]
```

```python
tfdif=TfidfVectorizer(ngram_range=(1,2),sublinear_tf=True,smooth_idf=True)
x_train=tfdif.fit_transform(x_train)
x_test=tfdif.transform(x_test)


print(x_train)
print("-------------------")
print(x_test)
```

I 🔄 split the dataset into training and testing sets to evaluate the model's performance on unseen data. Then, I applied 🏷️ **LabelEncoder** to the target column to convert categorical labels into numeric form suitable for machine learning algorithms. After that, I used ✍️ **TfidfVectorizer** to transform the cleaned text data into numerical feature vectors, capturing the importance of each word relative to the document and the entire corpus.


```python
model=SVC(C=8,kernel="linear",probability=True,class_weight="balanced")
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
accuracy=accuracy_score(y_test,y_pred)
print(accuracy)
```

```python
print(classification_report(y_test,y_pred))
```

```python
cm=confusion_matrix(y_test,y_pred)
cmd=ConfusionMatrixDisplay(cm)
cmd.plot()
```

```python
y_prob = model.predict_proba(x_test)[:, 1]

fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```
